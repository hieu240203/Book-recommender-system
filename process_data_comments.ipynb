```python
total_rating = data_comment.groupby('product_id')['rating'].sum()

comment_count = data_comment.groupby('product_id')['comment_id'].count()

average_rating = total_rating / comment_count

average_rating= round(average_rating,1)
```

```python
new_df = pd.DataFrame({'product_id': average_rating.index, 'average_rating': average_rating.values})
```
